```python
import pandas as pd
```

```python
data=pd.read_csv("nato_phonetic_alphabet.csv")
```

```python
phonetic_dict={row.letter:row.code for (index, row) in data.iterrows()}
phonetic_dict
```

```python
word=input("enter a word: ").upper()
word_letter=[phonetic_dict[letter] for letter in word]
print(word_letter)
```
